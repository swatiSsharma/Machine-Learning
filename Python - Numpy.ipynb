# 7.  Array Maths


```python
# creating the array

a = np.array([1,2,3,4,5,6])
b = np.array([7,8,9,10,11,12])
```

```python
# sum

print(a + b) # python ways of adding

np.add(a,b)
```

```python
# Subtraction/minus/difference

print(a-b)

np.subtract(a,b)
```

```python
# multiplication

print(a*b)

np.multiply(a,b)
```

```python
# division

print(a/b)

np.divide(a,b)
```

```python
 # square root
   
print(np.sqrt(a))

np.sqrt(b)

```

```python
# dot product and Matrix math

#NOTE: when we multiply in NumPy, we use elementwise multiplication and not matrix multiplication 

a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
c = np.array([2,3])
d = np.array([4,5])
```

```python
# inner product of vector

# through python

print(c.dot(d)) #

# through NumPy

np.dot(c,d) 

# NumPy - elementwise multiplication
#(2*4) + (3*5) --> 8 + 15 = 23

```

```python
# matrix *  vector product

print(a.dot(c))

np.dot(a,c)

# Numpy Elementwise multiplication
# ((1*2) + (2*3)) , ((3*2) + (4*3))
# (2+6),(6+12)
# 8,18

```

```python
a
```

```python
b
```

```python
# matrix by matrix product

print(a.dot(b))

np.dot(a,b)

# (1*5)+(2*7) ,(1*6)+(2*8)
# (3*5)+(2*7) ,(3*6)+(4*8)

# 5+14  , 5+16
# 15+14 , 18+32

# 19,22
# 43,50
```

# 10. Transpose - turn rows into columns and vice versa

```python
a = np.arange(4).reshape((2,2))
a
```
